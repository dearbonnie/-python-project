```python
class Person:
    def __init__(self,name):
        #建立一個attribute
        self.name = name

    #自訂實體的method
    def introduce(self):
        print(f"哈囉，我的名字是{self.name}")
```

```python
p3 = Person("jenny")
p3.introduce()
```

```python
p2= Person("rober")
p2.introduce()
```
