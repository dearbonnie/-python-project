```python
#學生總分為300分      
#有些學生可以加分5%     
while true:
    try:
        scores = int(input("請輸入學生分數(最高300分):"))
        break
    except valueError:
        print("輸入錯誤，請輸入整數!")

is_add = input("學生是否符合加分條件?(y,n)")

if is_add == "y":
    scores = scores * 1.05

scores = int(scores + 0.5)

if scores >300:
    scores = 300
    
print("學生分數:",scores)

```
