```python
#dictionary
#dictionary是key-value pair的資料結構，key是唯一的，value可以重複
#dictionary的key可以是任何不可變的資料類型，例如string、number、tuple等，value可以是任何資料類型，例如string、number、list、dictionary等
#dictionary的key-value pair是無序的，不能使用索引來存取資料，只能使用key來存取資料
#dictionary的key-value pair是可變的，可以新增、修改、刪除key-value pair
#dictionary的key-value pair是用大括號{}來表示，key-value pair之間用冒號:來分隔，key-value pair之間用逗號,來分隔
students = {
    '小明':80, 
    '小華': 90, 
    '小美': 85, 
    '小花': 95,
    '小王': 89}
print(students)
print(students['小明'])
print(students['小華'])
print(students['小美'])
print(students['小花'])
print(students['小王']) #索引從0開始


 #key-value pair
 #list
scores1 = [80,30,59,66,89]
for s in scores1:
    print(s)

#dict
students = {
    '小明':80,
    '小華': 90,
    '小美': 85,
    '小花': 95,
    '小王': 89}
for key in students:
    print(students[key])
```
